```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/drive/MyDrive/shai/Salaries.csv')
df.head()
```

```python
df.columns
```

```python
# Identify the number of rows and columns in the dataset
num_rows, num_columns = df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")

# Determine the data types of each column
data_types = df.dtypes
print("\nData Types:")
print(data_types)

# Check for missing values in each column
missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

```python
# Calculate basic statistics for the 'TotalPay' column
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode().iloc[0]  # In case there are multiple modes, pick the first one
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()
salary_range = max_salary - min_salary
std_dev_salary = df['TotalPay'].std()

# Display the results
print(f"Mean Salary: {mean_salary}")
print(f"Median Salary: {median_salary}")
print(f"Mode Salary: {mode_salary}")
print(f"Minimum Salary: {min_salary}")
print(f"Maximum Salary: {max_salary}")
print(f"Salary Range: {salary_range}")
print(f"Standard Deviation of Salary: {std_dev_salary}")
```

```python

# Data Cleaning: Handle missing data

# Handling missing values in numeric columns ('BasePay', 'OvertimePay', 'OtherPay', 'Benefits')
# You may choose to fill missing values with the mean of each respective column
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)

```

```python
df.drop(['Notes', 'Status'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

# Create a histogram for the distribution of salaries
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

```

```python
plt.figure(figsize=(10, 6))
department_counts = df['JobTitle'].value_counts()
top_departments = department_counts.head(10)  # Displaying the top 10 departments for better visualization
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

```python
# Group the data by 'JobTitle' and calculate summary statistics for each group
grouped_by_job_title = df.groupby('JobTitle')['TotalPay'].agg(['count', 'mean', 'min', 'max', 'std'])

# Display the summary statistics
print(grouped_by_job_title)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify correlation between 'TotalPay' and 'BasePay'
correlation = df['TotalPay'].corr(df['BasePay'])

# Print the correlation coefficient
print(f"Correlation between TotalPay and BasePay: {correlation}")

# Plot a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BasePay', y='TotalPay', data=df, alpha=0.7)
plt.title('Scatter Plot of TotalPay vs BasePay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.show()
```
